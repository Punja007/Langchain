```python
from langchain_community.document_loaders import WebBaseLoader
import bs4

webloader = WebBaseLoader(web_paths=('https://edition.cnn.com/2025/08/06/sport/kenny-bednarek-noah-lyles-sprinting-rivalry-intl',),
                          bs_kwargs=dict(parse_only=bs4.SoupStrainer(
                              class_=('headlineText', 'article__content')
                          ))
                          )
```

```python
web = webloader.load()
web
```
