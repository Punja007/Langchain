```python
from langchain_ollama import OllamaEmbeddings

embeddings = (OllamaEmbeddings(model='nomic-embed-text'))
```

```python
texts = [
    "The human brain contains approximately 86 billion neurons, each forming thousands of connections, making it more complex than any supercomputer.",
    "In 2025, artificial intelligence became mainstream in healthcare, automating diagnostics and assisting in personalized treatment plans.",
    "Photosynthesis is the process by which plants convert sunlight into chemical energy, producing oxygen as a by-product.",
    "LangChain allows developers to build applications powered by language models with memory, chaining, and external tools like search and databases.",
    "Black holes are regions of space where gravity is so strong that nothing, not even light, can escape from them."
]

embedded = embeddings.embed_documents(texts)
embedded
```

```python
len(embedded)
```

```python
len(embedded[0])
```

```python
query_result = embeddings.embed_query("by which process plants convert sunlight into chemical energy?")
```
